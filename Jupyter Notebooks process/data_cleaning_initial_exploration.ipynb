# Vanguard Digital Experiment — Data Cleaning & Initial Exploration

## Business Context

Vanguard launched an A/B test to evaluate whether a redesigned digital interface could improve user experience and increase process completion rates.

The experiment compares two groups:

- **Control Group:** clients who interacted with the traditional online process.
- **Test Group:** clients who interacted with the redesigned digital interface.

Both groups followed the same process flow:

`start → step_1 → step_2 → step_3 → confirm`

This notebook focuses on:

- Understanding the structure of the raw datasets.
- Identifying data quality issues.
- Cleaning and preparing the data.
- Creating a reliable dataset for EDA, KPI analysis, and hypothesis testing.

## 4. Validate Client Uniqueness

Before merging datasets, it is important to understand the level of granularity.

- Client profile data should contain one row per client.
- Experiment roster data should contain one row per client.
- Web data contains multiple rows per client because each row is an online event.

```python
# Check if client_id is unique in client profile and experiment datasets
uniqueness_check = pd.DataFrame({
    "Dataset": ["Client Profiles", "Experiment Roster", "Web Data Part 1", "Web Data Part 2"],
    "Rows": [df_demo.shape[0], df_exp.shape[0], df_web_1.shape[0], df_web_2.shape[0]],
    "Unique Clients": [
        df_demo["client_id"].nunique(),
        df_exp["client_id"].nunique(),
        df_web_1["client_id"].nunique(),
        df_web_2["client_id"].nunique()
    ]
})

uniqueness_check["One Row per Client?"] = uniqueness_check["Rows"] == uniqueness_check["Unique Clients"]
uniqueness_check
```

### Interpretation

If `Client Profiles` and `Experiment Roster` have one row per client, they can be safely merged with web data using `client_id`.

The web datasets are expected to have multiple rows per client because they represent behavioral events.

## 5. Concatenate Web Data

The two web data files are combined into one dataset to reconstruct the full digital journey.

```python
# Combine both web datasets
df_web = pd.concat([df_web_1, df_web_2], axis=0, ignore_index=True)

print("Combined web dataset shape:", df_web.shape)
print("Duplicated rows in combined web dataset:", df_web.duplicated().sum())

df_web.head()
```

## 6. Investigate Duplicated Web Events

Because web data represents client interactions, duplicated rows should not be removed without investigation.

A duplicate may represent:

- technical logging duplication
- page refresh
- repeated click
- accidental double action
- tracking issue

For this project, exact duplicated rows will be removed after documenting their volume, because they can inflate process step counts and distort journey metrics.

```python
# Display duplicated web rows
duplicated_web_rows = df_web[df_web.duplicated(keep=False)].sort_values(
    by=["client_id", "visitor_id", "visit_id", "date_time", "process_step"]
)

print("Total duplicated web rows:", df_web.duplicated().sum())
duplicated_web_rows.head(20)
```

```python
# Remove exact duplicated web events
df_web_clean = df_web.drop_duplicates().copy()

print("Web data before removing duplicates:", df_web.shape)
print("Web data after removing duplicates:", df_web_clean.shape)
print("Rows removed:", df_web.shape[0] - df_web_clean.shape[0])
```

## 7. Convert Date Columns

The `date_time` column is converted from object/text to datetime format.

This is necessary for:

- ordering user journeys
- calculating time between steps
- analyzing activity over time
- measuring process duration

```python
df_web_clean["date_time"] = pd.to_datetime(df_web_clean["date_time"])

df_web_clean.dtypes
```

## 8. Clean Client Profiles Dataset

The demographic dataset contains a small number of missing values.

Since missing values are limited, we first calculate the missing percentage to decide an appropriate strategy.

```python
# Missing value percentage in client profiles
missing_demo = pd.DataFrame({
    "missing_count": df_demo.isna().sum(),
    "missing_percentage": (df_demo.isna().mean() * 100).round(3)
}).sort_values("missing_count", ascending=False)

missing_demo
```

### Missing Data Strategy

Because the number of missing values is very small compared with the total dataset size, rows with missing critical demographic/account fields can be removed from demographic-based analysis.

For gender, unknown categories are preserved as `Unknown` because they may represent valid undisclosed information rather than missing data.

```python
# Create a clean copy
df_demo_clean = df_demo.copy()

# Standardize column values in gender
df_demo_clean["gendr"] = df_demo_clean["gendr"].replace({
    "U": "Unknown",
    "X": "Unknown"
})

# Drop rows with missing values because they represent a very small percentage of the dataset
df_demo_clean = df_demo_clean.dropna().copy()

print("Client profiles before cleaning:", df_demo.shape)
print("Client profiles after cleaning:", df_demo_clean.shape)
print("Rows removed:", df_demo.shape[0] - df_demo_clean.shape[0])

df_demo_clean["gendr"].value_counts(dropna=False)
```

## 9. Clean Experiment Roster Dataset

Clients without a Test or Control assignment cannot be used to evaluate the A/B test.

They will be excluded from experiment analysis while keeping the decision documented.

```python
# Check experiment assignment distribution
df_exp["Variation"].value_counts(dropna=False)
```

```python
# Keep only clients assigned to Test or Control
df_exp_clean = df_exp.dropna(subset=["Variation"]).copy()

print("Experiment roster before cleaning:", df_exp.shape)
print("Experiment roster after cleaning:", df_exp_clean.shape)
print("Rows removed:", df_exp.shape[0] - df_exp_clean.shape[0])

df_exp_clean["Variation"].value_counts()
```

## 10. Merge Cleaned Datasets

The cleaned web behavior dataset is merged with:

1. Experiment roster, to identify Test vs Control group.
2. Client profiles, to add demographic and account information.

An inner join with the experiment roster ensures that only users who were part of the A/B test are included in the analytical dataset.

```python
# Merge web data with experiment roster
df_merged = df_web_clean.merge(
    df_exp_clean,
    on="client_id",
    how="inner"
)

# Merge with client profiles
df_merged = df_merged.merge(
    df_demo_clean,
    on="client_id",
    how="left"
)

print("Final merged dataset shape:", df_merged.shape)
df_merged.head()
```

## 11. Final Data Quality Checks

After merging, we validate:

- missing values
- duplicated rows
- data types
- experiment group distribution
- process step values

```python
print("Final dataset shape:", df_merged.shape)

print("\nMissing values:")
display(df_merged.isna().sum().sort_values(ascending=False))

print("\nDuplicated rows:")
print(df_merged.duplicated().sum())

print("\nData types:")
display(df_merged.dtypes)

print("\nVariation distribution:")
display(df_merged["Variation"].value_counts())

print("\nProcess step distribution:")
display(df_merged["process_step"].value_counts())
```

## 12. Feature Engineering for Future Analysis

New variables are created to support EDA, Tableau, segmentation, KPI analysis, and storytelling.

Created features:

- `age_group`
- `tenure_group`
- `balance_segment`
- `year_month`
- `date`

```python
# Age groups
df_merged["age_group"] = pd.cut(
    df_merged["clnt_age"],
    bins=[0, 30, 50, 70, 100],
    labels=["Under 30", "30-50", "50-70", "70+"],
    include_lowest=True
)

# Tenure groups
df_merged["tenure_group"] = pd.cut(
    df_merged["clnt_tenure_yr"],
    bins=[0, 5, 10, 20, 70],
    labels=["New", "Established", "Loyal", "Veteran"],
    include_lowest=True
)

# Balance segments based on quartiles
df_merged["balance_segment"] = pd.qcut(
    df_merged["bal"],
    q=4,
    labels=["Low", "Medium", "High", "Very High"],
    duplicates="drop"
)

# Date-based features
df_merged["date"] = df_merged["date_time"].dt.date
df_merged["year_month"] = df_merged["date_time"].dt.to_period("M").astype(str)

df_merged.head()
```
